```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth
```

```python
df = pd.read_csv('../Data/basket_analysis.csv', index_col=0)
```

```python
freq_items_apriori = apriori(df, min_support=0.01, use_colnames=True)
```

```python
freq_items_apriori['length'] = freq_items_apriori['itemsets'].apply(lambda x: len(x))
```

```python
mask = freq_items_apriori['length'] > 1
freq_items_apriori[mask].sort_values('support', ascending=False)
```

```python
item_rules = association_rules(freq_items_apriori, metric='lift', min_threshold=1.2)
```

```python
item_rules.head()
```

```python
freq_items_fp = fpgrowth(df, min_support=0.01, use_colnames=True)
```

```python
freq_items_fp['length'] = freq_items_fp['itemsets'].apply(lambda x: len(x))
mask = freq_items_fp['length'] > 1
freq_items_fp[mask].sort_values('support', ascending=False)
```
